# README

Matplotlib is probably the Python's most popular library for data visualization. Matplotlib is a very flexible tools, allowing users to create and customize an impressive variety of plot 'types'. The best way to appreciate what you can achieve with Matplotlib is going through the [gallery of examples](https://matplotlib.org/gallery/index.html) reported in the website of the library. Personally, I tend to see these examples as a cookbook of plot 'recipes'. 

The [Bokeh library](https://bokeh.pydata.org/en/latest/) complements Matplotlib by offering some capabilities for interactive data visualization.

```python
# creating the figure
FIG = plt.figure(figsize=(4, 4))

# populate the figure with a plot
AX = FIG.add_subplot(1, 1, 1)

# creating the variables to plot
X = np.linspace(1, 10, 10)
Y = np.square(X)

# plot the data
AX.plot(X, Y, color="b", ls="--", marker="^")

# gridline
AX.grid(linestyle=":", linewidth=2)

# labels
AX.set_xlabel("X - axis")
AX.set_ylabel("Y - axis")

# title
AX.set_title("A title")

# save the figure
FIG.savefig("a_first_plot.pdf")

# display the figure
plt.show()

# close the FIG object
plt.close(FIG)
```

## Adding 'Textual' Elements to Plots

```python
# creating the figure
FIG = plt.figure(figsize=(4, 4))

# populate the figure with a plot
AX = FIG.add_subplot(1, 1, 1)

# creating the variables to plot
X = np.linspace(1, 10, 10)
Y = np.square(X)

# plot the data
AX.plot(X, Y, color="b", ls="--", marker="^")

# gridline
AX.grid(linestyle=":", linewidth=1)

# labels
AX.set_xlabel("X - axis")
AX.set_ylabel("Y - axis")

# title
AX.set_title("A title")

# annotations
AX.annotate("A simple comment", xy = (1, 90))

# tight layout
FIG.tight_layout()

# save the figure
# FIG.savefig("tight_layout.pdf")

# display the figure
plt.show()

# close the FIG object
plt.close(FIG)
```

```python
# creating the figure
FIG = plt.figure(figsize=(4, 4))

# populate the figure with a plot
AX = FIG.add_subplot(1, 1, 1)

# creating the variables to plot
X = np.linspace(1, 10, 10)
Y = np.square(X)
Y1 = np.sqrt(X)

# plot the data
AX.plot(X, Y, color="b", ls="", marker="^", label="Square")
AX.plot(X, Y1, color="red", ls="", marker="v", label="Square root")

# legend
AX.legend(loc="best")

# gridline
AX.grid(linestyle=":", linewidth=1)

# labels
AX.set_xlabel("X - axis")
AX.set_ylabel("Y - axis")

# title
AX.set_title("A title")

# annotations
# AX.annotate("A simple comment", xy = (1, 90))

# tight layout
FIG.tight_layout()

# save the figure
# FIG.savefig("tight_layout.pdf")

# display the figure
plt.show()

# close the FIG object
plt.close(FIG)
```

## Handling Subplots

```python
# creating the figure
FIG = plt.figure(figsize=(8, 4))

# populate the figure with a plot
# (num of rows, num of columns, position)
# position = 1 means first plot
AX0 = FIG.add_subplot(1, 2, 1)
AX1 = FIG.add_subplot(1, 2, 2)

# creating the variables to plot
X = np.linspace(1, 10, 10)
Y = np.square(X)
Y1 = np.sqrt(X)

# plot the data
AX0.plot(X, Y, color="b", ls="", marker="^", label="Square")
AX1.plot(X, Y1, color="red", ls="", marker="v", label="Square root")

# gridline
AX0.grid(linestyle=":", linewidth=1)
AX1.grid(linestyle=":", linewidth=1)

# labels
AX0.set_xlabel("X - axis")
AX0.set_ylabel("Y - axis")
AX1.set_xlabel("X - axis")
AX1.set_ylabel("Y - axis")

# tight layout
FIG.tight_layout()

# save the figure
# FIG.savefig("tight_layout.pdf")

# display the figure
plt.show()

# close the FIG object
plt.close(FIG)
```

## Types of Plots

### Histograms

```python
# creating a new figure
FIG = plt.figure(figsize=(4, 4))

# populate the figure with a plot
AX = FIG.add_subplot(1, 1, 1)

# simulate some data
np.random.seed(000)
X = np.random.randn(100)

# plot the data
AX.hist(X, bins=20)

# display the plot
plt.show()

# close the figure
plt.close(FIG)
```

```python
# creating a new figure
FIG = plt.figure(figsize=(4, 4))

# populate the figure with a plot
AX = FIG.add_subplot(1, 1, 1)

# simulate some data
np.random.seed(000)
X0 = np.random.randn(100)
X1 = np.random.normal(1, 1, 100)

# plot the data
AX.hist(X0, bins=20, color="b", alpha=0.20)
AX.hist(X1, bins=20, color="g", alpha=0.30)

# display the plot
plt.show()

# close the figure
plt.close(FIG)
```

### Scatter diagrams

```python
# create a figure
FIG = plt.figure(figsize=(4, 4))

# populate the figure with a plot
AX = FIG.add_subplot(1, 1, 1)

# create the data to plot
np.random.seed(000)
X0 = np.random.randn(100)
X1 = np.random.normal(1, 1, 100)

# plot the data
AX.scatter(X0, X1)

# display the plot
plt.show()

# close the figure
plt.close(FIG)
```

```python
# create a figure
FIG = plt.figure(figsize=(4, 4))

# populate the figure with a plot
AX = FIG.add_subplot(1, 1, 1)

# create the data to plot
np.random.seed(000)
X0 = np.random.randn(100)
X1 = np.random.normal(1, 1, 100)
Y = np.random.randint(1, 10, 100)
Z = np.random.poisson(12, 100) ** 2

# plot the data
AX.scatter(X0, X1, c=Y, s=Z, alpha=0.4)

# display the plot
plt.show()

# close the figure
plt.close(FIG)
```

### Boxplots

```python
# figure
FIG = plt.figure(figsize=(4, 4))

# plot
AX = FIG.add_subplot(111)

# plot the data
AX.boxplot([X0, X1])

# xtick labels
AX.set_xticklabels(("X0", "X1"))

# axhline
AX.axhline(y = 0, color = "r")

# display the plot
plt.show()

# close the figure
plt.close(FIG)
```

## From 2-D to 3-D Plots

```python
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
from matplotlib import cm

fig = plt.figure()
ax = fig.gca(projection='3d')
X, Y, Z = axes3d.get_test_data(0.05)

# Plot the 3D surface
ax.plot_surface(X, Y, Z, rstride=8, cstride=8, alpha=0.3)

# Plot projections of the contours for each dimension.  By choosing offsets
# that match the appropriate axes limits, the projected contours will sit on
# the 'walls' of the graph
cset = ax.contourf(X, Y, Z, zdir='z', offset=-100, cmap=cm.coolwarm)
cset = ax.contourf(X, Y, Z, zdir='x', offset=-40, cmap=cm.coolwarm)
cset = ax.contourf(X, Y, Z, zdir='y', offset=40, cmap=cm.coolwarm)

ax.set_xlim(-40, 40)
ax.set_ylim(-40, 40)
ax.set_zlim(-100, 100)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```
