# Problem 1

## Create a Pandas DF by reading the 'block_9.csv' file 

```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv('block_9.csv')
df1
```

## Create a new DF which reports the sum of reviews by year

```python
df1['reviews'] = 1
df2 = pd.DataFrame(df1.groupby('review_year')['reviews'].aggregate(np.sum))
df2
```

## Create a new DF which reports the sum of reviews by month and year

```python
df3 = pd.DataFrame(df1.groupby(['review_year', 'review_month'])['reviews'].aggregate(np.sum))
df3
```

## Create a new DF which reports the mean and standard deviation in the number of  reviews by month

```python
df4 = pd.DataFrame()
df4['mean'] = df3.groupby('review_month')['reviews'].aggregate(np.mean)
df4['std'] = df3.groupby('review_month')['reviews'].aggregate(np.std)
df4
```

# Problem 2

## Create a Pandas DF by reading the 'block_9.csv' file

```python
df5 = pd.read_csv('block_9.csv')
df5
```

## Create a new column reporting the length of the text of the review

```python
df5.info()
```

```python
df5['review_length'] = ' '
for index in df5.index:
    df5.loc[index, 'review_length'] = len(str(df5.loc[index, 'reviewText']))
df5
```

## Regress 'overall' on the newly created column (via OLS)

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
FML = 'overall ~ review_length'
OLS = smf.ols(FML, data = df5).fit()
print(OLS.summary())
```

## Make at least two Matplotlib plots showing the variability in the length of the review

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
FIG = plt.figure(figsize=(10, 5))
plot1 = FIG.add_subplot(1, 2, 1)
plot2 = FIG.add_subplot(1, 2, 2)
plot1.scatter(df5['timestamp'], df5['review_length'])
plot2.boxplot(df5['review_length'])
plt.show()
```
